```python
import matplotlib.pyplot as plt
import cv2
import glob
import os
import numpy as np
```

```python
images = glob.glob(os.path.join("./dataset_visualization/", "*png"))
```

```python
images_ = []
f, axarr = plt.subplots(2,6,figsize=(15,5))
for i, image in enumerate(images, 1):
    im = cv2.imread(image)
    im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
    images_.append(im)
    ax = plt.subplot(2, 6, i)
    ax.imshow(im)
    ax.axis('off')
    ax.set_title(f"Gesture {i}")
ax = plt.subplot(2, 6, 12)
ax.imshow(np.ones((225,225, 3))*255, cmap="gray")
ax.axis('off')
```

```python
for i in range(10):
    plt.imshow(images_[i])
    plt.show()
```
